```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# https://www.kaggle.com/datasets/rabieelkharoua/students-performance-dataset/data
```

```python
file="Student_performance_data _.csv"
df = pd.read_csv(file)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
df["Gender"] = df["Gender"].astype(str).replace("0", "Male").replace("1", "Female")


df["Ethnicity"] = df["Ethnicity"].astype(str)
ethnicity = {"0": "Caucasian", "1": "African American", "2": "Asian", "3": "Other"}
for i, e in df["Ethnicity"].items():
    df.loc[i, "Ethnicity"] = ethnicity[e]


df["ParentalEducation"] = df["ParentalEducation"].astype(str)
parental_education = {
    "0": "None",
    "1": "High School",
    "2": "Some College",
    "3": "Bachelor's",
    "4": "Higher",
}
for i,e in df['ParentalEducation'].items():
    df.loc[i, "ParentalEducation"] = parental_education[e]

df.head()
```

```python
# https://seaborn.pydata.org/examples/index.html
sns.set_theme(style="darkgrid")


sns.lineplot(data=df, x="Absences", y="GPA", hue="Extracurricular", palette="tab10")
```

```python
order = ["None", "High School", "Some College", "Bachelor's", "Higher"]
parental_ed = sns.barplot(
    data=df, x="ParentalEducation", y="GPA", palette="rocket", order=order
)
parental_ed.set(xlabel="Parental Education", ylabel="Average GPA")
```

```python
demo = sns.catplot(
    data=df, x="Age", y="GPA", hue="Gender", palette="rocket", kind="bar"
)
demo.set(xlabel="Age", ylabel="Average GPA")
```

```python
study = sns.lmplot(
    data=df,
    x="StudyTimeWeekly",
    y="GPA",
    #_bins=5,
    scatter=False,
    hue="ParentalSupport",
)
study.set(xlabel="Weekly Study Time", ylabel="Average GPA")
```
